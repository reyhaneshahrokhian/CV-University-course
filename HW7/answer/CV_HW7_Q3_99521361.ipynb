```python
import mtm
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
coin_image = image.imread('/content/coins.png')
plt.imshow(coin_image, cmap='gray')
```

```python
# Choose template
template = coin_image[109:145, 133:174]
plt.imshow(template, cmap="gray")
```

# Match templates

```python
# call matchTemplates
detections = mtm.matchTemplates(coin_image, [template], scoreThreshold=0.5, maxOverlap=0)

# Plot the detections
mtm.detection.plotDetections(coin_image, detections)
```

```python
# Another version to detect bigger coins better
big_template = coin_image[232:288, 18:75]
plt.imshow(big_template, cmap="gray")
```

```python
# call matchTemplates
detections = mtm.matchTemplates(coin_image, [template, big_template], scoreThreshold=0.45, maxOverlap=0)

# Plot the detections
mtm.detection.plotDetections(coin_image, detections)
```
